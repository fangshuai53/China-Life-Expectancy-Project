```python
#import lib
import pandas as pd
import numpy as np
import folium
from folium import FeatureGroup
import os
import branca
import vincent 
```

```python
#load and prepare dataset
CHN_geo = os.path.join('/Users/joned/Desktop/Github/China-Life-Expectancy-Project/DataSet/', 'CHN-provinces.json')
df0= os.path.join('/Users/joned/Desktop/Github/China-Life-Expectancy-Project/DataSet/', 'Data_1.xlsx')
CHN_life_expectancy_data = pd.read_excel(df0)

df1=CHN_life_expectancy_data.filter(['Chinese Name','1990 All','2000 All','2010 All'], axis=1).set_index('Chinese Name')
df2=CHN_life_expectancy_data.filter(['Chinese Name','1990 Female','2000 Female','2010 Female'], axis=1).set_index('Chinese Name')
df3=CHN_life_expectancy_data.filter(['Chinese Name','1990 Male','2000 Male','2010 Male'], axis=1).set_index('Chinese Name')
```

```python
#Analytic Popup vincent plot:
analysis1 = vincent.GroupedBar(df1)
analysis1.titles='预期寿命'
analysis1.axis_titles(x='地区', y='预期寿命（全体）')
analysis1.colors(brew='Set3')
analysis1.legend(title='年份')

analysis2 = vincent.GroupedBar(df2)
analysis2.titles='预期寿命'
analysis2.axis_titles(x='地区', y='预期寿命(女性)')
analysis2.colors(brew='Spectral')
analysis2.legend(title='年份')

analysis3 = vincent.GroupedBar(df3)
analysis3.titles='预期寿命'
analysis3.axis_titles(x='地区', y='预期寿命(男性)')
analysis3.colors(brew='Paired')
analysis3.legend(title='年份')
```

```python
#selectable layer (checkbox)
feature_group1=folium.FeatureGroup(name='1990年预期预期寿命信息',overlay=True)
feature_group2=folium.FeatureGroup(name='2000年预期预期寿命信息',overlay=True)
feature_group3=folium.FeatureGroup(name='2010年预期预期寿命信息',overlay=True)
feature_group4=folium.FeatureGroup(name='信息图表',overlay=True)

#base map
m=folium.Map(location=[42, 105], zoom_start=4)

#popup icons for life expectancy information by regions
for i in range(0,31):
    l1=folium.Marker([CHN_life_expectancy_data.iloc[i]['Latitude'], CHN_life_expectancy_data.iloc[i]['Longitude']], 
                     popup=folium.Popup(max_width=300,
                                        html=CHN_life_expectancy_data.iloc[i]['Chinese Name']+":"+'<br />' 
                     +"1990 预期寿命: "+np.str(CHN_life_expectancy_data.iloc[i]['1990 All'])+'<br />'
                     +"1990 预期寿命（男性）: "+np.str(CHN_life_expectancy_data.iloc[i]['1990 Male'])+'<br />'
                     +"1990 预期寿命（女性）: "+np.str(CHN_life_expectancy_data.iloc[i]['1990 Female'])),
                     icon=folium.Icon(color='blue')).add_to(feature_group1)

for i in range(0,31):
    l2=folium.Marker([CHN_life_expectancy_data.iloc[i]['Latitude'], CHN_life_expectancy_data.iloc[i]['Longitude']], 
                     popup=folium.Popup(max_width=300,
                                        html=CHN_life_expectancy_data.iloc[i]['Chinese Name']+":"+'<br />' 
                     +"2000 预期寿命: "+np.str(CHN_life_expectancy_data.iloc[i]['2000 All'])+'<br />'
                     +"2000 预期寿命（男性）: "+np.str(CHN_life_expectancy_data.iloc[i]['2000 Male'])+'<br />'
                     +"2000 预期寿命（女性）: "+np.str(CHN_life_expectancy_data.iloc[i]['2000 Female'])),
                     icon=folium.Icon(color='red')).add_to(feature_group2)

for i in range(0,31):
    l3=folium.Marker([CHN_life_expectancy_data.iloc[i]['Latitude'], CHN_life_expectancy_data.iloc[i]['Longitude']], 
                     popup=folium.Popup(max_width=300,
                                        html=CHN_life_expectancy_data.iloc[i]['Chinese Name']+":"+'<br />' 
                     +"2010 预期寿命: "+np.str(CHN_life_expectancy_data.iloc[i]['2010 All'])+'<br />'
                     +"2010 预期寿命（男性）: "+np.str(CHN_life_expectancy_data.iloc[i]['2010 Male'])+'<br />'
                     +"2010 预期寿命（女性）: "+np.str(CHN_life_expectancy_data.iloc[i]['2010 Female'])),
                     icon=folium.Icon(color='green')).add_to(feature_group3)

#popup incons for analytic results by years
popup1 = folium.Popup()
a1=folium.Vega(analysis1,height=550, width=1100).add_to(popup1)
folium.Marker([15, 100.51],icon=folium.Icon(color='orange',icon='align-left'),popup=popup1).add_to(feature_group4)

popup2 = folium.Popup()
a2=folium.Vega(analysis2,height=550, width=1100).add_to(popup2)
folium.Marker([15, 109.51],icon=folium.Icon(color='orange',icon='align-left'),popup=popup2).add_to(feature_group4)

popup3 = folium.Popup()
a3=folium.Vega(analysis3,height=550, width=1100).add_to(popup3)
folium.Marker([15, 91.51],icon=folium.Icon(color='orange',icon='align-left'),popup=popup3).add_to(feature_group4)


#layers of choropleth map of Chinese life expectancy (radio butt)
m1=m.choropleth(
     overlay=False,
     geo_data=CHN_geo,
     name='1990年中国人预期预期寿命',
     data=CHN_life_expectancy_data,
     columns=['Province Code', '1990 All'],
     key_on='feature.id',
     fill_color='YlOrRd',
     fill_opacity=0.7,
     line_opacity=0.2,
     legend_name='1990年中国人预期预期寿命')

m2=m.choropleth(
     overlay=False,
     geo_data=CHN_geo,
     name='2000年中国人预期预期寿命',
     data=CHN_life_expectancy_data,
     columns=['Province Code', '2000 All'],
     key_on='feature.id',
     fill_color='YlOrRd',
     fill_opacity=0.7,
     line_opacity=0.2,
     legend_name='2000年中国人预期预期寿命')

m3=m.choropleth(
     overlay=False,
     geo_data=CHN_geo,
     name='2010年中国人预期预期寿命',
     data=CHN_life_expectancy_data,
     columns=['Province Code', '2010 All'],
     key_on='feature.id',
     fill_color='YlOrRd',
     fill_opacity=0.7,
     line_opacity=0.2,
     legend_name='2010年中国人预期预期寿命')

m4=m.choropleth(
     overlay=False,
     geo_data=CHN_geo,
     name='1990年中国男性预期预期寿命',
     data=CHN_life_expectancy_data,
     columns=['Province Code', '1990 Male'],
     key_on='feature.id',
     fill_color='GnBu',
     fill_opacity=0.7,
     line_opacity=0.2,
     legend_name='1990年中国男性预期预期寿命')

m5=m.choropleth(
     overlay=False,
     geo_data=CHN_geo,
     name='2000年中国男性预期预期寿命',
     data=CHN_life_expectancy_data,
     columns=['Province Code', '2000 Male'],
     key_on='feature.id',
     fill_color='GnBu',
     fill_opacity=0.7,
     line_opacity=0.2,
     legend_name='2000年中国男性预期预期寿命')

m6=m.choropleth(
     overlay=False,
     geo_data=CHN_geo,
     name='2010年中国男性预期预期寿命',
     data=CHN_life_expectancy_data,
     columns=['Province Code', '2010 Male'],
     key_on='feature.id',
     fill_color='GnBu',
     fill_opacity=0.7,
     line_opacity=0.2,
     legend_name='2010年中国男性预期预期寿命')

m7=m.choropleth(
     overlay=False,
     geo_data=CHN_geo,
     name='1990年中国女性预期预期寿命',
     data=CHN_life_expectancy_data,
     columns=['Province Code', '1990 Female'],
     key_on='feature.id',
     fill_color='PuRd',
     fill_opacity=0.7,
     line_opacity=0.2,
     legend_name='1990年中国女性预期预期寿命')

m8=m.choropleth(
     overlay=False,
     geo_data=CHN_geo,
     name='2000年中国女性预期预期寿命',
     data=CHN_life_expectancy_data,
     columns=['Province Code', '2000 Female'],
     key_on='feature.id',
     fill_color='PuRd',
     fill_opacity=0.7,
     line_opacity=0.2,
     legend_name='2000年中国女性预期预期寿命')

m9=m.choropleth(
     overlay=False,
     geo_data=CHN_geo,
     name='2010年中国女性预期预期寿命',
     data=CHN_life_expectancy_data,
     columns=['Province Code', '2010 Female'],
     key_on='feature.id',
     fill_color='PuRd',
     fill_opacity=0.7,
     line_opacity=0.2,
     legend_name='2010年中国女性预期预期寿命')


feature_group1.add_to(m)
feature_group2.add_to(m)
feature_group3.add_to(m)
feature_group4.add_to(m)

folium.LayerControl().add_to(m)

m.save('China 1990, 2000, 2010 life expectancy (CN).html')
```
